# Excercise - 01 

## :- |  3 Channel (RGB) Image Manipulation Using Matplotlib

```python
# Import Library
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define the Path
```

```python
path = "../data-warehouse/images/sample/"
```

```python
#Read the Image File
```

```python
img = plt.imread(path + "rgb-one.jpeg");
```

```python
#Check the Shape of the Image
```

```python
img.shape
```

```python
# Check the type of img instance
```

```python
type(img)
```

```python
# Show the Image
```

```python
plt.imshow(img);
```

```python
# Get the Red Channel and Print the Image out of It
```

```python
red_channel = img[:, :, 0]
plt.imshow(red_channel, cmap='Reds')
```

```python
# Get the Green Channel and Print the image out of it
```

```python
green_channel = img[:, :, 1]
plt.imshow(green_channel, cmap='Greens')
```

```python
# Get the Blue Channel and Print the image out of It
```

```python
blue_channel = img[:, :, 2]
plt.imshow(blue_channel, cmap='Blues')
```
